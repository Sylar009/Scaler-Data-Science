```python
def convertToMinutes(time):
    return time.seconds/60
```

```python
data['DURATION_MINS'] = (data['END_DATE'] - data['START_DATE']).apply(convertToMinutes)
```

```python
data.head()
```

```python
data['SPEED_KMPH']= (data['KMs']/(data['DURATION_MINS']/60)).round(1)
```
